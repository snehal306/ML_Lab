```python
# import libraries
import numpy as np
import pandas as pd

# import the data required
data = pd.read_csv("/content/drive/My Drive/Study/ML/Dataset2.csv")
print(data.head())
```

```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['Outlook', 'Temp','Wind','Humidity','Class'])
print(one_hot_encoded_data)
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test= train_test_split(one_hot_encoded_data, test_size=0.25, random_state=94)
data_train.head()
```

**Q1 - What will be the value of Play, if Outlook is ’Rainy’, Temperature is ’Mild’, Humidity =’Nor-

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
#Create a Classifier
model=MultinomialNB()
# Train the model using the training sets
model.fit(data_train.iloc[: , :-2],data_train["Class_Y"])
data_train.iloc[: , :-2]
```

```python
#Predict Output
predicted= model.predict([[0,1,0,0,0,1,1,0,0,0,1]])
print(predicted)
```

**Q3- Accuracy , precision and recall**

```python
#import the necessary module
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
#Train the model using the training sets
gnb.fit(data_train,data_train.Class_Y )
#Predict the response for test dataset
target_pred = gnb.predict(data_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(data_test.Class_Y, target_pred))
```

```python
#Import confusion_matrix from scikit-learn metrics module for confusion_matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(data_test.Class_N, target_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(data_test.Class_Y, target_pred, average=None)
recall = recall_score(data_test.Class_Y, target_pred, average=None)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```
