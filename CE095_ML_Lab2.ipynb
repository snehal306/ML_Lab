```python
#feature selection
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('/content/Exercise-CarData.csv')
data = data.iloc[:,:-1]
data.head()
data.info()
```

```python
#select feature based on corelation
#1)Generating the co-relation matrix
corr = data.corr()
corr.head()
```

```python
#2)Generating the co-relation heatmap
sns.heatmap(corr)
```

```python
#3)Now, will we compare the correlation between features and remove one of two features that have a correlation higher than 0.9
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```
